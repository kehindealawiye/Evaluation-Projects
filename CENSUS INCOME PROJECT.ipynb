## Step 3: Data Preprocessing


### 3.1: Fixing '?' values in the dataset

```python
census = census.replace('?', np.nan)
```

```python
# Checking null values
round((census.isnull().sum() / census.shape[0]) * 100, 2).astype(str) + ' %'
```

```python
columns_with_nan = ['Workclass', 'Occupation', 'Native_country']
```

```python
for col in columns_with_nan:
    census[col].fillna(census[col].mode()[0], inplace=True)
```

### 3.2: Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in census.columns:
    if census[col].dtypes == 'object':
        encoder = LabelEncoder()
        census[col] = encoder.fit_transform(census[col])
```

### 3.3: Feature Selection

```python
X = census.drop('Income', axis=1)
Y = census['Income']
```

```python
from sklearn.ensemble import ExtraTreesClassifier
selector = ExtraTreesClassifier(random_state=42)
```

```python
selector.fit(X, Y)
```

```python
feature_imp = selector.feature_importances_
```

```python
for index, val in enumerate(feature_imp):
    print(index, round((val * 100), 2))
```

```python
X.info()
```

```python
X = X.drop(['Workclass', 'Education', 'Race', 'Sex',
            'Capital_loss', 'Native_country'], axis=1)
```

### 3.4: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
for col in X.columns:
    scaler = StandardScaler()
    X[col] = scaler.fit_transform(X[col].values.reshape(-1, 1))
```

### 3.5: Fixing imbalanced dataset using Oversampling

```python
round(Y.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
```

```python
ros.fit(X, Y)
```

```python
X_resampled, Y_resampled = ros.fit_resample(X, Y)
```

```python
round(Y_resampled.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

### 3.6: Creating a train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_resampled, Y_resampled, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

## Step 6: Hyperparameter Tuning
